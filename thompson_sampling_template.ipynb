# Thompson Sampling Template

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
```

## Importing the dataset
- expects a click-through rate type dataset with row representing each training sample, column with binary digits representing either an item is not picked or is picked (0 or 1)

```
dataset = pd.read_csv('ENTER_THE_NAME_OF_YOUR_DATASET_HERE.csv')
```

## Implementing Thompson Sampling
- rows represent the number of rows, or the numbers of training samples provided in the dataset; defaulted to full row lengths
- columns represent the number of columns, or the numbers of items exist in each row; defaulted to full column lengths

```
rows = len(dataset.values)
columns = len(dataset.values[0])
items_selected = []
num_rewards_1 = [0] * columns
num_rewards_0 = [0] * columns
total_reward = 0
for row in range(0, rows):
    item = 0
    max_random = 0
    for column in range(0, columns):
        random_beta = random.betavariate(num_rewards_1[column] + 1, num_rewards_0[column] + 1)
        if random_beta > max_random:
            max_random = random_beta
            item = column
    items_selected.append(item)
    reward = dataset.values[row, item]
    if reward == 1:
        num_rewards_1[item] += 1
    else: 
        num_rewards_0[item] += 1
total_reward = num_rewards_1
```

## Visualising the results - Histogram

```
plt.hist(items_selected)
plt.title('Histogram of Item Selection')
plt.xlabel('Items')
plt.ylabel('Number of times each item was selected')
plt.show()
```
