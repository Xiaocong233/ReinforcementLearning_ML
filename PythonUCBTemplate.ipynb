# Upper Confidence Bound Template

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

## Importing the dataset
- expects a click-through rate type dataset with row representing each training sample, column with binary digits representing either an item is not picked or is picked (0 or 1)

```
dataset = pd.read_csv('ENTER_THE_NAME_OF_YOUR_DATASET_HERE.csv')
```

## Implementing UCB
- rows represent the number of rows, or the numbers of training samples provided in the dataset; defaulted to 10000
- columns represent the number of columsn, or the numbers of items exist in each row; defaulted to 10

```
rows = 10000
columns = 10
items_selected = []
frequencies_of_selections = [0] * columns
cumulative_rewards = [0] * columns
total_rewards = 0

for row in range(0, rows):
    item = 0
    max_upper_bound = 0
    for column in range(0, columns):
        if frequencies_of_selections[column] > 0:
            avg_reward = cumulative_rewards[column] / frequencies_of_selections[column]
            margin_of_error = math.sqrt(3/2 * math.log(row + 1) / frequencies_of_selections[column])
            upper_bound = avg_reward + margin_of_error
        else:
            upper_bound = 1e400
        if upper_bound > max_upper_bound:
            max_upper_bound = upper_bound
            item = column
    items_selected.append(item)
    frequencies_of_selections[item] += 1
    reward = dataset.values[row, item]
    cumulative_rewards[item] += reward
total_rewards = sum(cumulative_rewards)
```

## Visualising the results

```
plt.hist(items_selected)
plt.title('Histogram of Item Selection')
plt.xlabel('Items')
plt.ylabel('NUmber of times each item was selected')
plt.show()
```
